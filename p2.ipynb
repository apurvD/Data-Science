```python
mean=df['raisedhands'].mean()
mean
```

```python
std_dev=df['raisedhands'].std()
std_dev
```

```python
lower=mean-3*std_dev
lower
```

```python
upper=mean+3*std_dev
upper
```

```python
lowest=df['raisedhands'].min()
lowest
```

```python
highest=df['raisedhands'].max()
highest
```

```python
#z-score
zs=(df['raisedhands']-mean)/std_dev
zs
```

```python
zs.describe().T
```

```python
zsoutliers=[]
for i in zs:
    if (i>3 or i<-3):
        zsoutliers.append(i)
```

```python
zsoutliers
```

```python
lower
```

```python
upper
```

```python
outliers=[]
for i in df['raisedhands']:
    if (i>upper or i<lower):
        outliers.append(i)
```
