```python
print("The minimum value of Age =",df['Age'].min())
print("The minimum value of Years of Experience =",df["YearsExperience"].min())
print("The minimum value of Salary =",df['Salary'].min())

print("The maximum value of Years of Experience =",df["YearsExperience"].max())
print("The maximum value of Salary =",df['Salary'].max())
print("The maximum value of Age =",df['Age'].max())

print("The mean of Age =",df['Age'].mean())
print("The mean of Years of Experience =",df["YearsExperience"].mean())
print("The mean of Salary =",df['Salary'].mean())

print("The median of Age =",df['Age'].median())
print("The median of Years of Experience =",df["YearsExperience"].median())

print("The median of Salary =",df['Salary'].median())
print("The mode of Age =",df['Age'].mode())
print("The Standard deviation of Age =",df["Age"].std())
print("The Standard deviation of Years of Experience =",df["YearsExperience"].std())
print("The Standard deviation of Salary =",df["Salary"].std())
```

```python
#grouping the salaries with number of people in that particular age
df["Salary"].groupby(df["Age"]).count()
```

```python
#finding out the maximum salary of a particular age
df["Salary"].groupby(df["Age"]).max()
```
